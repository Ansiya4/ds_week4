```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
# Load the Iris dataset
iris = load_iris(as_frame=True)
df = iris.data
```

```python
# Specify the number of clusters (K)
k = 3
```

```python
# Initialize and fit the KMeans model
kmeans = KMeans(n_clusters=k)
kmeans.fit(df)
```

```python
# Add cluster labels to the DataFrame
df['Cluster'] = kmeans.labels_
df['Cluster']
```

```python
print(df)
```

```python
# Plot the data points with different colors for each cluster
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=df['Cluster'], cmap='rainbow')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('K-Means Clustering (Sepal)')
plt.show()
```

```python
# Evaluate clustering using silhouette score
silhouette_avg = silhouette_score(df.iloc[:, :-1], df['Cluster'])
print("Silhouette Score:", silhouette_avg)
```

```python
# Evaluate clustering using Davies-Bouldin index
davies_bouldin_avg = davies_bouldin_score(df.iloc[:, :-1], df['Cluster'])
print("Davies-Bouldin Index:", davies_bouldin_avg)
```
