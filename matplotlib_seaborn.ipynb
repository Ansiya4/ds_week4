```python
plt.hist(flower_df.sepal_width,bins=10)
```

```python
import numpy as np
plt.hist(flower_df.sepal_width,bins=np.arange(2,5,.25));
```

```python
plt.hist(flower_df.sepal_width,bins=[1,3,4,4.5])
```

```python
setosa_df = flower_df[flower_df.species=='setosa']
versicolor_df = flower_df[flower_df.species=='versicolor']
virginica_df = flower_df[flower_df.species=='virginica']

plt.hist(setosa_df.sepal_width,alpha=.4,bins=np.arange(2,5,.25),stacked=True)
plt.hist(versicolor_df.sepal_width,alpha=.4,bins=np.arange(2,5,.25),stacked=True)
plt.hist(virginica_df.sepal_width,alpha=.4,bins=np.arange(2,5,.25),stacked=True)
plt.legend(['setosa','versicolor','virginica'])
```

```python
plt.hist([setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width],alpha=1,bins=np.arange(2,5,.25),stacked=True)
plt.legend(['setosa','versicolor','virginica'])
```
